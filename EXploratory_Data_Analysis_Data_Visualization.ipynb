```
import pandas as p
import numpy as n
f=p.read_csv('/content/train.csv')
```

```
import matplotlib.pyplot as plt
import seaborn as s
```

Countplot


```
s.countplot(x='Survived',data=f,hue='Sex')
```

```
s.countplot(x='Survived',data=f)
```

Line chart

```
s.lineplot(x='Age',y='Fare',data=f)
```

Scatter Plot

```
f.plot.scatter('Age','Fare')
```

```
f.plot.scatter('Age','Survived')
```

Histogram 

```
f['Age'].hist(bins=8)
```

Pie Charts

```
s= f['Survived'].value_counts()
fig1,ax1 = plt.subplots()
ax1.pie(s,labels=['Not Survived','Survived'],autopct='%1.1f%%',shadow=True)
plt.show()
```
