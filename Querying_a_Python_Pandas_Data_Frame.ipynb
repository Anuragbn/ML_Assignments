```
import pandas as p
import numpy as n
f=p.read_csv('/content/train.csv')
```

Projecting a column i.e. a feature 

```
f['Name'].head()
```

Selecting rows and columns

```
f.loc[0:15,'Name':'Embarked']
```

```
f.iloc[0:15,0:4]
```

Indexing and Row Selection

Print the name of the oldest passenger travelling on ship

```
f[f['Age']==f['Age'].max()]['Name']
```

Sorting

```
f.sort_values(by='Age',ascending='True').head()
```

```
f.sort_values(by=['Name','Age'],ascending=['True','False']).head()
```

Replacing values in column
1. Through map() method


```
n={'male':0,'female':1}
f['Sex']=f['Sex'].map(n)
f.head()
```

  2. Through replace method


```
o={0:'male',1:'female'}
f=f.replace({'Sex':o})
f.head()
```

GroupBy

```
f.groupby(by='Survived')['Age'].describe()
```

Contingency Table: Relating two variables


```
p.crosstab(f['Sex'],f['Survived'], normalize=True)
```

```
f.pivot_table(['Age', 'Fare'],['Survived'], aggfunc='mean')
```
